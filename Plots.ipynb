```python
# load the data
d_types=[('periodID',object),('period_initial_goal_differential',int),('blocked_shots_away',int),\
         ('blocked_shots home',int),('faceoff_away',int),('faceoff_home',int),('giveaway_away',int),\
         ('giveaway_home',int),('hit_away',int),('hit_home',int),('missed_shot_away',int), \
         ('missed_shot_home',int),('penalty_away',int),('penalty_home',int),('shot_away',int),\
        ('shot_home',int),('takeaway_away',int),('takeaway_home',int),('period_final_goal_differential',int)]

data = np.genfromtxt('data/periodStats.txt',delimiter=',',dtype=d_types,skip_header=1)

data.shape
```

```python
# relationshp between fields and period_final_goal_differential

sns.set(rc={'figure.figsize':(4,4)})
sns.palplot(sns.color_palette("Paired"))

period_final_goal_diff = data['period_final_goal_differential']
period_initial_goal_diff = data['period_initial_goal_differential']
away_blocked = data['blocked_shots_away']
home_blocked = data['blocked_shots_home']
away_faceoff = data['faceoff_away']
home_faceoff = data['faceoff_home']
away_giveaway = data['giveaway_away']
home_giveaway = data['giveaway_home']
away_hit = data['hit_away']
home_hit = data['hit_home']
away_missed_shot = data['missed_shot_away']
home_missed_shot = data['missed_shot_home']
away_penalty = data['penalty_away']
home_penalty = data['penalty_home']
away_shot = data['shot_away']
home_shot = data['shot_home']
away_takeaway = data['takeaway_away']
home_takeaway = data['takeaway_home']

idx = np.random.choice(np.arange(len(away_blocked)), 10000, replace=False)

peiod_goal_diff = np.subtract(period_final_goal_diff,period_initial_goal_diff)

blocked_differential = np.subtract(home_blocked,away_blocked)
x_sample = blocked_differential[idx]
y_sample = peiod_goal_diff[idx]
jnt = sns.jointplot(x=x_sample, y=y_sample,kind="kde",color='blue',height=4)
jnt.set_axis_labels('Blocked Differential', 'Goal Differential',color='blue' ,fontsize=16)
plt.show()

b = sns.distplot(blocked_differential)
b.set_title("Blocked Differential Distribution",color='blue',fontsize=16)
plt.show()

faceoff_differential = np.subtract(home_faceoff,away_faceoff)
x_sample = faceoff_differential[idx]
y_sample = peiod_goal_diff[idx]
jnt = sns.jointplot(x=x_sample, y=y_sample,kind="kde",color='green',height=4);
jnt.set_axis_labels('Faceoff Differential', 'Goal Differential', color='green',fontsize=16)
plt.show()

b = sns.distplot(faceoff_differential,color='green')
b.set_title("Faceoff Differential Distribution",color='green',fontsize=16)
plt.show()

giveaway_differential = np.subtract(home_giveaway,away_giveaway)
x_sample = giveaway_differential[idx]
y_sample = peiod_goal_diff[idx]
jnt = sns.jointplot(x=x_sample, y=y_sample,kind="kde",color='orange',height=4);
jnt.set_axis_labels('Giveaway Differential', 'Goal Differential', color='orange' ,fontsize=16)
plt.show()

b = sns.distplot(giveaway_differential,color='orange')
b.set_title("Giveaway Differential Distribution",color='orange',fontsize=16)
plt.show()

hit_differential = np.subtract(home_hit,away_hit)
x_sample = hit_differential[idx]
y_sample = peiod_goal_diff[idx]
jnt = sns.jointplot(x=x_sample, y=y_sample,kind="kde",color='red',height=4);
jnt.set_axis_labels('Hit Differential', 'Goal Differential', color='red', fontsize=16)
plt.show()

b = sns.distplot(hit_differential,color='red')
b.set_title("Hit Differential Distribution",color='red',fontsize=16)
plt.show()

missed_shot_differential = np.subtract(home_missed_shot,away_missed_shot)
x_sample = missed_shot_differential[idx]
y_sample = peiod_goal_diff[idx]
jnt = sns.jointplot(x=x_sample, y=y_sample,kind="kde",color='purple',height=4);
jnt.set_axis_labels('Missed Shot Differential', 'Goal Differential', color='purple', fontsize=16)
plt.show()

b = sns.distplot(missed_shot_differential,color='purple')
b.set_title("Missed Shot Differential Distribution",color='purple',fontsize=16)
plt.show()

penalty_differential = np.subtract(home_penalty,away_penalty)
x_sample = penalty_differential[idx]
y_sample = peiod_goal_diff[idx]
jnt = sns.jointplot(x=x_sample, y=y_sample,kind="kde",color='brown',height=4);
jnt.set_axis_labels('Penalty Differential', 'Goal Differential', color='brown', fontsize=16)
plt.show()

b = sns.distplot(penalty_differential,color='brown')
b.set_title("Penalty Differential Distribution",color='brown',fontsize=16)
plt.show()

shot_differential = np.subtract(home_shot,away_shot)
x_sample = shot_differential[idx]
y_sample = peiod_goal_diff[idx]
jnt = sns.jointplot(x=x_sample, y=y_sample,kind="kde",color='yellow',height=4);
jnt.set_axis_labels('Shot Differential', 'Goal Differential', color='yellow', fontsize=16)
plt.show()

b = sns.distplot(shot_differential,color='yellow')
b.set_title("Shot Differential Distribution",color='yellow',fontsize=16)
plt.show

takeaway_differential = np.subtract(home_takeaway,away_takeaway)
x_sample = takeaway_differential[idx]
y_sample = peiod_goal_diff[idx]
jnt = sns.jointplot(x=x_sample, y=y_sample,kind="kde",color='pink',height=4);
jnt.set_axis_labels('Takeaway Differential', 'Takeaway Differential', color='pink', fontsize=16)
plt.show()

b = sns.distplot(takeaway_differential,color='pink')
b.set_title("Takeaway Differential Distribution",color='pink',fontsize=16)
plt.show()
```

```python
#violin plot
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler(feature_range=(-1, 1), copy=True)

blocked_differential = minmax.fit_transform(blocked_differential.reshape(-1,1))
faceoff_differential = minmax.fit_transform(faceoff_differential.reshape(-1,1))
giveaway_differential = minmax.fit_transform(giveaway_differential.reshape(-1,1))
hit_differential = minmax.fit_transform(hit_differential.reshape(-1,1))
missed_shot_differential = minmax.fit_transform(missed_shot_differential.reshape(-1,1))
penalty_differential = minmax.fit_transform(penalty_differential.reshape(-1,1))
shot_differential = minmax.fit_transform(shot_differential.reshape(-1,1))
takeaway_differential = minmax.fit_transform(takeaway_differential.reshape(-1,1))
period_final_goal_diff = minmax.fit_transform(period_final_goal_diff.reshape(-1,1))
period_goal_diff = minmax.fit_transform(peiod_goal_diff.reshape(-1,1))

np.column_stack((blocked_differential,faceoff_differential,giveaway_differential,hit_differential,missed_shot_differential,\
                penalty_differential,shot_differential,takeaway_differential))

df = pd.DataFrame({'blocked_differential': blocked_differential[:, 0], \
                   'faceoff_differential': faceoff_differential[:, 0], \
                  'giveaway_differential': giveaway_differential[:,0], \
                  'hit_differential': hit_differential[:,0], \
                  'missed_shot_differential': missed_shot_differential[:,0], \
                  'penalty_differential': penalty_differential[:,0], \
                  'shot_differential': shot_differential[:,0], \
                  'takeaway_differential': takeaway_differential[:,0], \
                  'period_goal_diff': period_goal_diff[:,0]})

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 6))

# Draw a violinplot with a narrower bandwidth than the default
v = sns.violinplot(data=df, palette="Set3", bw=.2, cut=1, linewidth=1)
v.set_title("Violin Plot - All Attributes and Outcome",color='black',fontsize=16)

# Finalize the figure
ax.set(ylim=(-1, 1))

sns.despine(left=True, bottom=True)
plt.xticks(rotation=90)
plt.show()
```
